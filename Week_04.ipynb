### Monday: Lists
In other languages they are called arrays and arrays have similar characteristics as lists.
A list is a data structure in Python that is a mutable, ordered sequence of elements. Mutable means that you can change the 
items inside, while ordered sequence is in reference to index location. The first element in a list will always be located at
index 0
Each element in a list is called an item.
Data collections are simply data types that can store multiple items.

```python
###Declaring Lists
#Defining a list is like any other data type; on the left of the operator is the name of the variable, and on the
# right is the value. The difference here is that the value is a set of items declared between square brackets.
#To separate each item within a list, we simply use commas. Let’s try:

nums = [9, 15.4, 6.7, 8]
print(nums)

#We use an index to access an item value from a list
print(nums[1]) #will output the value at index 1 = 15.4
num = nums[2] #saves index value 2 into num
print(num) #prints a value assigned to  num

#Declaring a list of mixed data types
num = 4.5
data =[num, "word", True]  #The power of data collection
print(data)

#Lists within lists: How to store lists within another list

data1 = [5, "book", [34, "hello"], True]
print(data1)
print(data1[2])


#Accessing elements of a list within a list we use bracket notation as shown bellow

print(data1[2][0])#outputs 34 which is an element in the second list within the data1 list

inner_list = data1[2]#inner_list is [34, "hello"]
print(inner_list[1])#prints "hello" from the innerlist

#Changing values in a list
# When you work with lists you need to be able to alter the value of the items within the
# list. It’s like re-declaring a normal variable to a different value, except you access the index first:

data2 = [5,67,89,6,5]
print(data2)# it outputs the list of data2 before altering its elements
data2[0]=100
print(data2)#prints the output of the elements of data2 variable after altering

#Variable Storage
#when a variable is declared the elements are stored in a memory location and the location
#can be accessed by aspecific reference ID the id() function.
#Example

a= [54, 873, 87]
print(id(a))#large number represents a  memory location

# When a list is stored in memory, each item is given its own location. Changing the
# value using index notation will change the value stored within that memory block. Now,
# if a variable’s value is another variable, like so:

b=a

# Changing the value at a specific index will change the value for both lists. Let’s see an
# example:
# understanding how lists are stored

print("a: {}\t b: {}".format(a,b))
print("Location a[0]: {}\t Location b[0]: {}".format(id(a[0]), id(b[0])))
a[0]=20 #re-declaring the value of a[0] also changes b[0]   
print("a: {}\t b: {}".format(a,b))

#Copying a list: You create a similar list without altering the original list by copying.
#using [:] to copy a list
data3 = [5,10,15,20]
data_copy = data3[:] # a single colon copies
print("data: {}\t data_copy: {}".format(data3, data_copy))

########You can also use the method .copy( ).#########
```

### Tuesday: For Loops

```python
# Rather than writing the same lines of code several times, we use loops. In Python there
# are two types of loops, today’s lesson will be on “For Loops.”
# Loops are how programmers rerun the same lines of code several times. Loops will
# always run until a condition is met.
# Note It’s always important to condense your code down to as few lines as
# possible, as it is more efficient for the program.

# For loops are primarily used to loop a set number of times.
# Everyf or loop begins with the keyword “for”. Then you define a temporary variable, sometimes
# known as a counter or index. Next is the “in” keyword, followed by the range function
# (which will be explained later). Lastly, we have a colon to end the statement
# All for loops will follow this exact structure of keyword, variable, keyword, function, and colon

# writing your first for loop using range


for num in range(5):
    print("Value: {}".format(num)) #This loop is essentially counting to five and printing out each number
    
# Note The value 5 is not output because range( ) counts up to but not including

# Range allows us to count from one number to another while being able to define where
# to start and end and how much we increment or decrement by. Meaning that we could
# count every other number or every fifth number if we wanted to. When used with a for
# loop, it gives us the ability to loop a certain number of times.

#providing the start, stop, and step for the range function
#This example prints all the evens between 2 and 10

print("\n start, stop, step \n")
for num in range(2, 10, 2):
    print("Value: {}".format(num))

# Looping by Element
# When working with data types that are iterable, meaning they have a collection of
# elements that can be looped over, we can write the for loop differently:

# printing all characters in a name using the 'in' keyword

name = "John Smith"
for letter in name:
    print("Value: {}".format(letter))

    
#Statements that can be used with the for loop:
# 1. Continue Statement
#Once a continue statement is hit, the current iteration stops and goes back to the top of the loop
#Example: using the continue statement within a foor loop

print("\nContinue Statement\n")

for num in range(5):
    if num == 3:
        continue
    print(num)
    
# The output will result in “0, 1, 2, 4” because the continue
# statement is only read when num is equal to the value of 3. Once the statement is hit,
# it stops the current iteration and goes back to the top to continue looping on the next
# iteration. This completely stops the code below continue from being interpreted, so it
# doesn’t hit the print statement
    
#Break Statement
#It allows us to break out of a loop at any point in time. Let’s see an example:
#Breaking out of a loop using the "break" keyword

print("\nBreak Statement\n")

for num in range(5):
    if num == 3:
        break
    print(num)

# The output will result in “0, 1, 2” because we broke the
# loop completely when num was equal to 3. Once a break is read, the loop completely
# stops, and no more code within the loop is run

# If you use a double loop, the break statement will only break out of the loop
# that the statement is within. Meaning, it will not break out of both loops if the break
# statement is used within the inner loop.

## Pass Statement: The pass statement is simply just a placeholder
# so that the program doesn’t break. Let’s see an example:

#setting a placeholder using the 'pass' keyword

print("\n Pass Statemnt\n")

for i in range(5):
   # TODO: #Add code to break number
        pass

# Go ahead and run that cell. Nothing happens, but that’s a good thing. If you take the
# pass statement out completely, the program will break because there needs to be some
# sort of code within the block

############### Tuesday Exercises ##################
# Divisible by Three: Write a for loop that prints out all numbers from 1 to 100
# that are divisible by three.

print("The following is a list of numbers between 0-100 divisible by 3")
for i in range(100):
    if i % 3 == 0: #Use the modulus operator to find the remainder. If the remainder is 0 then the number is divisible by 3
        print(i)



# Only Vowels: Ask for user input, and write a for loop that will output all the
# vowels within it. For example:
# >>> "Hello" ➔ "eo"

#Also this works
# ans = input("Enter a word: ")

# for letter in ans:
#     if letter in ['a', 'e', 'i', 'o', 'u']:
#         print(letter)

vowel = "aeiouAEIOU"
word = str(input("Enter a word: "))
print("The vowels in the word" +word+ " are: ")
for vowel in word:
    if(vowel == 'a' or vowel == 'A' or vowel == 'e' or vowel == 'E' or
      vowel == 'i' or vowel == 'I' or vowel == 'o' or vowel == 'O' or
      vowel == 'u' or vowel == 'U'):
        
        print(str(vowel))
```

### Thursday: Working with Lists

```python
# Checking length

print("Checking the lenght of the list\n")
nums = [17, 24, 98, 46]
length = len(nums) # len() returns an integer, the lenght of the list i.e. 4
print(length)

# Slicing lists
# Slicing follows the same arguments as the range function start, stop, step:
# accessing specific items of a list with slices

print("\nSlicing\n")
print(nums[1:3]) # will output items in index 1 and 2
print(nums[:2]) # will output items in index 0 and 1
print(nums[::2]) # # will print every other index - 0, 2, 4, etc.
print(nums[-2:]) # will output the last two items in list

# Adding Items
# When you need to add items to your lists, Python has two different methods for doing so.
# .append()--- Append will always add the value within the parenthesis to the back of the list. Let’s see:

print("\nAdd Items to the List\n")
# adding an item to the back of a list using append

print("\n.append()\n")
nums = [10,20,30]
nums.append(45)
print(nums)

# .insert( )
# The second method to add items to a list is using insert. This method requires an index
# to insert a value into a specific location. Let’s see an example:

print("\n.insert()\n")
sentence = ["Keep", "love", "safe"]
sentence.insert(1, "the")
print(sentence)

print("\nRemoving Items From a List\n")
# These are the two main methods of removing items from a list

# .pop()
# By default, the pop method removes the last item in the list; however, you can specify an
# index to remove as well. This method is also widely used to save the removed item too.
# When pop is used, it not only removes the item but also returns it. This allows us to save
# that value into a variable to be used later:

print("\n.Pop()\n")
# using pop to remove items and saving to a variable to use later

items = [5, "This", True]
items.pop() # by default removes the last item
removed_item = items.pop(0) # removes 5 and saves it into the variable
print(removed_item, "\n", items)

# .remove( )
# The remove method allows us to remove items from a list based on their given value:

print(".remove()")
sports = ["Baseball", "Hockey", "Football", "Handball", "Volleyball"]
try:
    sports.remove("Football")
except:
    print("That item does not exist in the list")
print(sports)


############ Working with numerical lists #################
# max, min and sum functions are some of the functions used frequently in python

# using min, max and sum
print("\nWorking with Numerical Lists \n")
print("\nMin, Max, and Sum\n")
numss = [10, 19, 67, 5]
print("The maximum number is {}".format(max(numss))) # Finds and prints the maximum number in the list
print("The minimum number is {}".format(min(numss))) # Finds and prints the minimum number in the list
print("The sum of the numbers is {}".format(sum(numss))) # Finds and prints all the sum in the list


########### Sorting a list #############
# .sorted() -----> The sorted function will work on either numerical or alphabetical lists, but not one that is mixed
# using sorted on lists for numerical and alphabetical data
print("\nSorting a list: .sorted()\n")
numsx = [5, 8, 0, 2]
sorted_nums = sorted(numsx) #save to a new variable to use later
print(numsx, sorted_nums) # the original list is in tact

# .sort( )
# The sort method is used for the same purpose that our previous sorted function is used
# for; however, it will change the original list directly:
print("\n.sort()\n")
numsx.sort() # alters the list directly
print(numsx)

############ Conditionals and Lists ############
# Using “in” and “not in” Keywords
# When working with lists, they serve a purpose to find values within the list quickly:
print("\nConditionals and Lists\n")
names = ["Mary", "Jack", "James"]
if "Mary" in names:
    print("Found") # will run since Mary is in the list
if "Kim" not in names:
    print("Not Found") # will run since Jimmy is not in the list

############# Checking an Empty List ###############
# It’s usually to ensure you don’t cause any errors in your program, so let’s see how we can check:
# using conditionals to see if a list is empty
print("\nChecking an Empty List\n")
if names == []: # Could also say 'if not names'
    print("Empty")
else:
    print("Not empty")


############# Loops and Lists ############
# You can use both the for and while loops to iterate over the items within a list.

#### for loops ###
# When iterating over a list with a for loop, the syntax looks like when we used the range
# function previously; however, this time we use a temporary variable, the in keyword,
# and the name of the list. For each iteration, the temporary variable is assigned the item’s
# value. Let’s try it out:

# using a for loop to print all items in a list

print("\nLoops an Lists\n")
print("\nFor Loop\n")
sports = ["Baseball", "Hockey", "Football", "Handball", "Volleyball"]
for sport in sports:
    print(sport)
    
###### Using While Loops #######

print("\nWhile Loop\n")
# using the while loop to remove a certain value
names = ["Mary", "Jack", "James", "Jane", "Jude"]
while "Jack" in names:
    names.remove("Jack")
print(names)
```

### Friday: Creating Hangman Project

### Final Design
As always, we want to lay out our final design before we begin coding. This week will
not be based around graphics, like last week, so we’ll focus on the logic and the steps
necessary to run the program. Luckily for us, the logic is essentially the steps needed to
play the game:

1. Select a word to play with.

2. Ask user for input.

3. Check if guess is correct

a. If it is, show the letter in the proper place.

b. If it isn’t, lose a life.

4. Continue steps 2 and 3 until one of the following occurs:

a. The user guesses the word correctly.

b. The user loses all their lives.

This is the main game play functionality. There are several other steps we need to
perform before actually running the game, like declaring game variables; however, this
is the primary functionality that we needed to lay out before we begin coding. Knowing
this structure will allow us to stay on track with our program

```python
# Add additional imports
from random import choice #choice will selsct a random item
from IPython.display import clear_output # clear_output used to clear output so that it does not continue outputting on top of each other

# Daclaring game variables
words = ["tree", "basket", "chair", "paper", "python"]
word = choice(words) # randomly choose a words list
guessed, lives, game_over = [], 7, False # multi variable assignment

# create a list of underscores to the lenght of the word
guesses = ["_"] * len(word) # we enable the user to see the length of the word

# Create main game loop: The first step is  to be able to accept user input and stop playing the game
while not game_over:
    ans = input("Type quit or guess a letter: ").lower()
    #output game information
    hidden_word = "".join(guesses)
    print("Your guessed letters: {}".format(guessed)) # outputting guessed letters
    print("Word to guess: {}".format(hidden_word))
    print("Lives: {}".format(lives))
    ans = input("Type quit or guess a letter").lower()
    clear_output() # Clear all previous output
    if ans == "quit":
        print("Thanks for playing.")
        game_over = True 
    elif ans in word and ans not in guessed: #check if letter in word
        print("You guessed correctly!")
        #create a loop to change underscore to proper letter
        for i in range(len(word)):
            if word[i] == ans: #compares values at indexes
                guesses[i] = ans
    elif ans in guessed:
        print("You already guessed that. Try again.")
    else: #otherwise lose life
        lives -= 1
        print("Incorrect you lost a life")
        if ans not in guessed:
            guessed.append(ans) #add answer to guessed list
        if lives <=0:
            print("You lost all your lives, you lost")
            game_over = True
        #winning condition
        elif word == "".join(guesses):
            print("Congratulations, you guessed it correctly!")
            game_over = true


```
