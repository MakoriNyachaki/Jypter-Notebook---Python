### Wednesday: Working with Strings

```python
#String concatenation is adding one string to another string using the addition sign without using variables
#Example
#concatenating strings while using the addition operator without variables

name = "Jane" + " " + "Rock"
print(name)

#concatenating strings while using the addition operator with variables
fName = "Sweet"
lName = "Potatoes"
vname = fName + " " + lName
print(vname)
# print(fName + " " + lName)#you can also use this to print the names.

##String formatting allows us to write an entire string and inject the variables we want to use in the proper locations

#######   .format( )   #######
# The format method works by putting a period directly after the ending string quotation,
# followed by the keyword “format”. Within the parenthesis after the keyword are the
# variables that will be injected into the string
#use of curly brackets come into play to place the variables in the correct location in the string

# injecting variables using the format method
name = "John"
print("Hello {}".format(name))
print("Hello {}! You are {} yrs old.".format(name, 28))

#### f Strings in python 3.6 or higher ######
# The new way to inject variables into a string in Python is by using what we call f strings.
# By putting the letter “f” in front of a string, you’re able to inject a variable into a string
# directly in line. This is important, as it makes the string easier to read when it gets longer,
# making this the preferred method to format a string.
#example

print(f"Hello {name}")

# Python 2 doesn’t include the .format() method; instead you would use percent
# operators to mark the location of the variable being injected. The following is an
# example to inject the variable “name” into the location of “%s”. The letter after the
# percent operator signifies the data type. For integers, you would use “%d” for digit.
# After the string closes, you would place a percent operator, followed by the variables
# you would like to use. Let’s look at an example:

## one major difference between versions 2 & 3

print('Hello, %s' %name)

# python 2 multiple variable formatting
print("Hello,%s %s" %(fName, lName)) #surround the variables in parenthesis

##### String Indexing  ##########
#indexing starts at 0 in most programming languages including pyhton
word = "Hello"
print(word[0])#prints "H"
print(word[1])#prints "e"
print(word[-1])#prints "o",the last letter of the string 'Hello'

# Be very careful when working with indexes. An index is a specific location in
# memory. If you try to access a location that is out of range, you will crash your program
# because it’s trying to access a place in memory that does not exist. For example, if we
# tried to access index 5 on the “Hello”.

####  String Slicing #####
# Slicing is used mostly with Python
# lists; however, you can use it on strings as well. Slicing is essentially when you only want
# a piece of the variable, such that if I only wanted “He” from the word “Hello”, we would
# write the following:

print(word[0:2]) #outputs 'He'

##  variable_name[ start : stop : step ] ##
# In the previous slicing example, we only included the start and stop because the step is optional
# and defaults to incrementing by one each time. However, what if we wanted to print
# every other letter
# By passing the step as the number two, it increments
# the index by two each time instead of one. We will cover this more in depth in a later
# chapter; for now let this be an introduction into slicing with all three arguments.
#Example

print(word[0:5:2])#outputs 'Hlo'

###  Wednesday Exercise  ###

#variable injection
# Create a print statement that injects an integer,
# float, boolean, and string all into one line. The output should look like
# “23 4.5 False John”.

inte = 23
flo = 4.5
boo = False
name = "John"

print("{} {} {} {}".format(inte, flo, boo, name))

#Fill in the blanks
# Using the format method, fill in the following
# blanks by assigning your name and favorite activities into variables:
# "{ }'s favorite sports is { }."
# "{ } is working on { } programming!"

myName = "Makotech"
mySport = "Badminton"
myWorking = "Python"

print("{}'s favorite sport is {}".format(myName, mySport))
print("{} is working on {} programming". format(myName, mySport))




```

### Thursday: String Manupulation

```python
####  .title()  ####

#The title method capitalizes all first letters in each word of a string. Example:

name = "star chalk"
cname = "COW DUNG"
print(name.title())#capitalizes the string - first letter of each word on the string
print(name.upper())# capitalizes the whole string - each word in the string
print(cname.lower())#outputs the string in lowercase - the whole string

###  .replace() ###
# The replace method works like a find and replace tool. It takes in two values within its
# parenthesis, one that it searches for and the other that it replaces the searched value with:

print(name.replace("star", "board"))

# For the replace to be stored properly afterward, we would have to re-declare
# our words variable: words = words.replace(‘!’, ‘.’).

###  .find()  ###
#The find method will search for any string we ask it to. Example:

# finding the starting index of our searched term

print(name.find("chalk"))

### .strip()   ###
# used to get rid of a certain character on the left and right side of a string
# By default, it will remove spaces

#Example: removing white space with strip

names = "    John    "
print(names.strip())
print(names.rstrip())#removes white space charcaters to the right only
print(names.lstrip())#removes whitespace characters to the left of the string

###  .split()   ###
# split simply returns a list and we haven’t covered those quite yet. Let us see how it works

s ="I love mum!"
print(s.split())#outputs "['I', 'love', 'mum!']"


### Thursday Exercises ###

##Uppercasing
#Try manipulating the string “uppercase” so it prints out as all 
# uppercase letters. You’ll need to look up a new method.

stru = "Dad travelled yesterday"
print(stru.upper())#.upper() method for uppercase and .lower() method for lowercase

##Strip symbols
# Strip all the dollar signs from the left side of this string “$$John
# Smith”. Try it with .lstrip( ) and .strip( ). To see a description on how to use
# the strip method further, try using the help function in Python by typing the
# following:
# >>> help(" ".strip)

stre = "$$John Smith"
print(stre.lstrip("$")) #Removes the $ signs to the left of the string
print(stre.rstrip("$")) #Removes the $ signs to the right of the string
print(stre.strip("$")) #strips $ signs both to the left and right of the string


#####for more string manipulation methods in python visit:
######## >>>>>>https://www.w3schools.com/python/python_ref_string.asp>>>>   #########
```
